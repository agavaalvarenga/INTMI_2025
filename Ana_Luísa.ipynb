```python
def eh_primo(numero):
    """Verifica se um número é primo."""
    if numero <= 1:
        return False

    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False
    return True

numeros_originais = []
print("Digite 10 números inteiros:")
for _ in range(10):
    num = int(input("Digite um número: "))
    numeros_originais.append(num)
numeros_primos = []
for numero in numeros_originais:
    if eh_primo(numero):
        numeros_primos.append(numero)
print("\n--- Resultados ---")
print(f"Lista original: {numeros_originais}")
print(f"Números primos encontrados: {numeros_primos}")

```

```python
def verificar_palindromo_lista(lista):
    """Verifica se uma lista de 8 elementos é palíndroma."""
    lista_invertida = lista[::-1]
    if lista == lista_invertida:
        return True
    else:
        return False
lista_de_elementos = []
print("Digite 8 elementos para a lista:")
for i in range(8):
    elemento = input(f"Elemento {i+1}: ")
    lista_de_elementos.append(elemento)
if verificar_palindromo_lista(lista_de_elementos):
    print("\nA lista é um palíndromo!")
else:
    print("\nA lista NÃO é um palíndromo.")

```

```python
matriz = []
for i in range(3):
    linha = []
    for j in range(3):
        valor = float(input(f"Digite o valor da posição [{i}][{j}]: "))
        linha.append(valor)
    matriz.append(linha)
a, b, c = matriz[0]
d, e, f = matriz[1]
g, h, i = matriz[2]
det = (a * (e*i - f*h)
       - b * (d*i - f*g)
       + c * (d*h - e*g))
print("\nMatriz digitada:")
for linha in matriz:
    print(linha)

print("\nDeterminante =", det)
```

```python
print("Digite os valores da primeira matriz 3x3:")
matriz1 = []
for i in range(3):
    linha = []
    for j in range(3):
        valor = float(input(f"1ª matriz [{i}][{j}]: "))
        linha.append(valor)
    matriz1.append(linha)
print("\nDigite os valores da segunda matriz 3x3:")
matriz2 = []
for i in range(3):
    linha = []
    for j in range(3):
        valor = float(input(f"2ª matriz [{i}][{j}]: "))
        linha.append(valor)
    matriz2.append(linha)
soma = []
diferenca = []

for i in range(3):
    linha_soma = []
    linha_dif = []
    for j in range(3):
        linha_soma.append(matriz1[i][j] + matriz2[i][j])
        linha_dif.append(matriz1[i][j] - matriz2[i][j])
    soma.append(linha_soma)
    diferenca.append(linha_dif)
print("\nMatriz 1:")
for linha in matriz1:
    print(linha)

print("\nMatriz 2:")
for linha in matriz2:
    print(linha)

print("\nMatriz Soma:")
for linha in soma:
    print(linha)

print("\nMatriz Diferença:")
for linha in diferenca:
    print(linha)
```

```python
print("Digite 10 elementos para o conjunto A:")
A = set()
for i in range(10):
    valor = input(f"A[{i+1}]: ")
    A.add(valor)
print("\nDigite 10 elementos para o conjunto B:")
B = set()
for i in range(10):
    valor = input(f"B[{i+1}]: ")
    B.add(valor)
intersecao = A & B
uniao = A | B
diferenca_AB = A - B
diferenca_BA = B - A
print("\nConjunto A:", A)
print("Conjunto B:", B)

print("\nInterseção (A ∩ B):", intersecao)
print("União (A ∪ B):", uniao)
print("Diferença A - B:", diferenca_AB)
print("Diferença B - A:", diferenca_BA)
```

```python
import math
print("Digite números inteiros (separe por espaço):")
conjunto = set(map(int, input().split()))

quadrados_perfeitos = set()

for n in conjunto:
    if n >= 0:
        raiz = int(math.sqrt(n))
        if raiz * raiz == n:
            quadrados_perfeitos.add(n)

print("\nConjunto original:", conjunto)
print("Quadrados perfeitos encontrados:", quadrados_perfeitos)
```

```python
dicionario = {}

for i in range(10):
    palavra = input(f"Digite a {i+1}ª palavra: ").strip()
    primeira_letra = palavra[0].lower()

    if primeira_letra not in dicionario:
        dicionario[primeira_letra] = []

    dicionario[primeira_letra].append(palavra)

print("\nDicionário organizado por primeira letra:")
for letra, palavras in dicionario.items():
    print(f"{letra}: {palavras}")
```

```python
produtos = {}
for i in range(3):
    nome = input(f"Digite o nome do produto {i+1}: ")

    precos = []
    for j in range(3):
        preco = float(input(f"Digite o preço do produto '{nome}' na loja {j+1}: R$ "))
        precos.append(preco)

    produtos[nome] = precos
maior_variacao = -1
produto_maior_var = ""

for nome, precos in produtos.items():
    variacao = max(precos) - min(precos)
    if variacao > maior_variacao:
        maior_variacao = variacao
        produto_maior_var = nome
print("\nProdutos e preços:")
for nome, precos in produtos.items():
    print(f"{nome}: {precos}")

print(f"\nProduto com maior variação de preço: {produto_maior_var}")
print(f"Variação: R$ {maior_variacao:.2f}")
```

```python
frase = input("Digite uma frase: ").lower()
pontuacao = ".,;:!?()[]{}\"'"
for p in pontuacao:
    frase = frase.replace(p, "")

palavras = frase.split()
contagem = {}
for palavra in palavras:
    if palavra in contagem:
        contagem[palavra] += 1
    else:
        contagem[palavra] = 1

print("\nNúmero de ocorrências de cada palavra:")
for palavra, qtd in contagem.items():
    print(f"{palavra}: {qtd}")
```
